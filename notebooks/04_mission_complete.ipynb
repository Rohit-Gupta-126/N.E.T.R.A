```python
from eodag import EODataAccessGateway

# Initialize NETRA (Letting it choose the best path automatically)
print("üöÄ NETRA Connecting...")
dag = EODataAccessGateway()

# The Search
print("Searching for Sentinel-2 data...")
results = dag.search(
    productType='S2_MSI_L1C',
    start='2024-01-01',
    end='2024-01-03',
    geom={'lonmin': 85.8, 'latmin': 20.2, 'lonmax': 85.9, 'latmax': 20.3}
)

if len(results) > 0:
    print(f"‚úÖ MISSION SUCCESS! NETRA found {len(results)} images.")
    print(f"Data Source: {results[0].provider}") 
    print(f"Image ID: {results[0].properties['title']}")
    print("------------------------------------------------")
    print("We have established contact with Global Satellites.")
    print("Proceeding to Phase 2: ISRO Integration.")
else:
    print("‚ùå Critical Failure. No internet or data access.")
```
