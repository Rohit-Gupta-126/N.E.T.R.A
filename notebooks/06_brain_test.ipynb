```python
import sys
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv(dotenv_path="../.env")

# 1. Setup Path
sys.path.append(os.path.abspath("../src"))

# 2. Inject Credentials from .env (The Engine Fuel)
os.environ["GEMINI_API_KEY"] = os.getenv("GEMINI_API_KEY")
# ESA Credentials
os.environ["EODAG__COPERNICUS_DATASPACE__AUTH__CREDENTIALS__USERNAME"] = os.getenv("EODAG__COPERNICUS_DATASPACE__AUTH__CREDENTIALS__USERNAME")
os.environ["EODAG__COPERNICUS_DATASPACE__AUTH__CREDENTIALS__PASSWORD"] = os.getenv("EODAG__COPERNICUS_DATASPACE__AUTH__CREDENTIALS__PASSWORD")
os.environ["EODAG__COPERNICUS_DATASPACE__PRIORITY"] = os.getenv("EODAG__COPERNICUS_DATASPACE__PRIORITY")

# ISRO Credentials
os.environ["BHOONIDHI_USER"] = os.getenv("BHOONIDHI_USER")
os.environ["BHOONIDHI_PASS"] = os.getenv("BHOONIDHI_PASS")

# 3. Load the Brain
from graph.workflow import create_netra_graph

print("üöÄ Booting up NETRA Unified Agent...")
netra_brain = create_netra_graph()

# 4. Run the Mission
initial_state = {
    "query": "Find satellite images of New York City for December 2024",
    "parameters": {},
    "results": [],
    "final_response": "",
    "errors": []
}

print("\n--- MISSION START ---")
final_state = netra_brain.invoke(initial_state)
print("--- MISSION COMPLETE ---\n")

# 5. Report
print(f"üìä Total Images Found: {len(final_state['results'])}")
print("-" * 30)
for img in final_state['results']:
    print(f"üåç Source: {img['source']}")
    print(f"ID: {img['id']}")
    print("-" * 30)

if final_state['errors']:
    print("\n‚ö†Ô∏è ERRORS ENCOUNTERED:")
    for err in final_state['errors']:
        print(f"- {err}")
```
